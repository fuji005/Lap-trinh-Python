```python
"TUPLE"

numbers = (1,2,3,3,4,5,6,7,7)

# 1. create a tuple of numbers and print one item.
print(numbers[0])

# 2. to unpack a tuple into several variables.
(a,b,c,d,e,f,g,h,i) = numbers
print(f"b = {b}")

# 3. to add an item to a tuple.
print(f"Before: {numbers}")
x = list(numbers)
x.append(8)
numbers = tuple(x)
print(f"After: {numbers}")

# 4. to find the index of an item in a tuple.
print(f"Vi tri cua so 4: {numbers.index(4)}")

# 5. to find the repeated items of a tuple
repeated = []
for x in numbers:
    y = numbers.count(x)
    if y > 1:
        if x not in repeated:
            print(f"{x} is repeated {y} times")
            repeated.append(x)
```

```python
"SET"

# 1. Write a Python program to find the maximum and minimum values in a set.
my_set = {1, 2, 3, 4, 5, 6}
max
min
for x in my_set:
    if x > max:
        max = x
    if x < min:
        min = x
print(f"Max = {max}")
print(f"Min = {min}")

# 2. Write a Python program to check if a given value is present in a set or not.
given_value = 8
if given_value in my_set:
    print("\nYes")
else: print("\nNo")

# 3. Write a Python program to check if two given sets have no elements in common.
set_1 = {1, 2, 3, 4, 5, 6}
set_2 = {7, 8, 9, 10}
if set_1.intersection(set_2):
    print("\nHave elements in common")
else: print("\nHave no elements in common")

# 4. Write a Python program to find all the unique words and count the frequency of occurrence from a given list of strings. Use Python set data type.
list_string = ["apple banana apple", "banana orange apple", "grape banana kiwi"]

unique_words = set()

word_count = {}

# Duyệt qua từng chuỗi trong danh sách
for sentence in list_string:
    for word in sentence.split():
        word = word.lower()
        unique_words.add(word)

        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1

print("\nTần suất xuất hiện:")
for word in unique_words:
    print(f"{word}: {word_count[word]}")

# 5. Given two sets of numbers, write a Python program to find the missing numbers in the second set as compared to the first and vice versa. Use the Python set
set_3 = {1,2,3,4,5}
set_4 = {4,5,6,7,8}
print(f"\nElements of set 3 that not present on set 4: {set_3.difference(set_4)} ")
print(f"Elements of set 4 that not present on set 3: {set_4.difference(set_3)} ")
```
